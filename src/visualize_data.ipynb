```python
import IPython.display as display
```

```python
from read_records import read_full_tf_training_records, _parse_image_function
```

```python
dataset_train_full = read_full_tf_training_records()
parsed_image_dataset = dataset_train_full.map(_parse_image_function)
print(parsed_image_dataset)
```

```python
for image_features in parsed_image_dataset.take(2):
    print(f"image_name: {image_features['image_name'].numpy().decode('utf-8')}")
    print(f"eye: {image_features['eye'].numpy()}")
    print(f"level_cat: {image_features['level_cat'].numpy()}")
    print("image:")
    display.display(display.Image(data=image_features['image'].numpy()))
    print("\n\n")
```
